```python
count=1
delivery_df=pd.DataFrame()
for index,row in matches.iterrows():
    if count in [75,100,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1
    ball_of_match=[]
    batsman=[]
    bowler=[]
    runs=[]
    player_of_dismissed=[]
    teams=[]
    batting_team=[]
    match_id=[]
    city=[]
    venue=[]
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    loop_df=pd.DataFrame({
        'match_id':match_id,
        'teams':teams,
        'batting_team':batting_team,
        'ball':ball_of_match,
        'batsman':batsman,
        'bowler':bowler,
        'runs':runs,
        'player_dismissed':player_of_dismissed,
        'city':city,
        'venue':venue
    })
    delivery_df=delivery_df._append(loop_df)
```

```python
output['runs'] = pd.to_numeric(output['runs'], errors='coerce')  # 'coerce' will handle non-numeric values by converting them to NaN
```

```python
output['current_score'] = output.groupby('match_id')['runs'].cumsum()
```

```python
output
```

```python
#focus on balls and wickets left
#separating overs and balls
output['over']=output['ball'].apply(lambda x:str(x).split(".")[0])
output['ball_no']=output['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
output
```

```python
output['balls_bowled']=(output['over'].astype('int')*6)+output['ball_no'].astype('int')
```

```python
output
```

```python
output['balls_left']=300- output['balls_bowled']
output['balls_left'] = output['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
print(output[output['player_dismissed'] == '0'])
```

```python
output['player_dismissed'] = output['player_dismissed'].apply(lambda x: 0 if x == '0' else 1)

# Calculate cumulative sum of dismissals within each match
output['player_dismissed'] = output.groupby('match_id')['player_dismissed'].cumsum()

# Calculate wickets left
output['wickets_left'] = 10 - output['player_dismissed']

# Ensure 'wickets_left' doesn't go below 0
output['wickets_left'] = output['wickets_left'].clip(lower=0)
```

```python
output
```

```python
output['crr'] = (output['current_score']*6)/output['balls_bowled']
```

```python
output
```

```python
groups = output.groupby('match_id')

match_ids = output['match_id'].unique()
last_five = []

for id in match_ids:
    group = groups.get_group(id)
    # Ensure 'runs' is numeric using .loc
    group.loc[:, 'runs'] = pd.to_numeric(group['runs'], errors='coerce')
    
    # Apply rolling sum only on 'runs' column
    rolling_sum = group['runs'].rolling(window=30).sum()
    
    last_five.extend(rolling_sum.values.tolist())
```

```python
output['last_five'] = last_five
```

```python
final_df = output.groupby('match_id').sum()['runs'].reset_index().merge(output,on='match_id')
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
